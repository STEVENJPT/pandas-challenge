```python
import pandas as pd
import numpy as np
```

```python
file = "PandasHomework.csv"
```

```python
data= pd.read_csv(file)
data.head()
```

data.head()

```python
pre_number_players = data["SN"].value_counts()
number_players = len(pre_number_players)
players_dicts={"Total Players": str(number_players)}
players_frame=pd.DataFrame([players_dicts])
players_frame
```

```python
items=len(data["Item ID"].unique())
average_price = data["Price"].mean()
purchases= data["Purchase ID"].count()
revenue = data["Price"].sum()
Summary_df=pd.DataFrame([{"Number of Unique Items": str(items), "Average Price": str(average_price), "Number of Purchases": str(purchases), "Total Revenue": str(revenue)}])
Summary_df["Average Price"]=Summary_df["Average Price"].astype(float).map("{:.2f}".format)
Summary_df["Total Revenue"]=Summary_df["Total Revenue"].astype(float).map("${:,.2f}".format)
Summary_df
```

```python
data_gender = data.loc[:,["SN", "Gender"]]
data_gender = data_gender.drop_duplicates()
gender_group_df= data_gender["Gender"].value_counts()
gender_group = pd.DataFrame(gender_group_df)
gender_group["Share of players"]= gender_group["Gender"] / number_players*100
gender_group["Share of players"]=gender_group["Share of players"].astype(float).map("{:,.1f}%".format)
gender_group.head(10)
```

```python
data_by_gender = pd.DataFrame(data.groupby("Gender")["Purchase ID"].count())
data_by_gender["Avg_Price"] = pd.DataFrame(data.groupby("Gender")["Price"].mean())
data_by_gender["Total Purchase Value"] = pd.DataFrame(data.groupby("Gender")["Price"].sum())
data_by_gender["Average Purchase"] = pd.DataFrame(data.groupby("Gender")["Price"].sum() / gender_group_df)
data_by_gender["Average Purchase"]= data_by_gender["Average Purchase"].astype(float).map("{:,.2f}".format)
data_by_gender.head()
```

```python
data1 = data.drop_duplicates("SN", keep="first")
bins = pd.cut(data1['Age'], [0,9, 14,19,24,29,34,39,100])
data_age = pd.DataFrame(data1.groupby(bins)['Age'].count())
data_age = data_age.rename(columns={"Age":"Total Count"})
data_age["Percentage of Total"] = data_age["Total Count"]/data_age["Total Count"].sum()*100
data_age["Percentage of Total"] = data_age["Percentage of Total"].astype(float).map("{:,.1f}%".format)
data_age
```

```python
data_spend = pd.DataFrame(data.groupby("SN")['Price'].count())
data_spend = data_spend.rename(columns={"Price":"Purchase Count"})
data_spend["Average Purchase Price"] = pd.DataFrame(data.groupby("SN")['Price'].mean())
data_spend["Total Purchase Value"] = pd.DataFrame(data.groupby("SN")['Price'].sum())
data_spend= data_spend.sort_values("Total Purchase Value", ascending=False)
data_spend.head(5)
```
